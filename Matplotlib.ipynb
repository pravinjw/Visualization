## Matplotlib Pyplot

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Scatter Chart

```python
day = [1,2,3,4,5,6,7]
temp = [30,33,34,32,28,27,32]
plt.scatter(day,temp,alpha=0.8)
plt.xlabel("Days")
plt.ylabel("Temp")
plt.title("Day and Temp")
plt.show()
```

### Line Chart

```python
# Single graph
day = [1,2,3,4,5,6,7]
temp = [30,33,34,32,28,27,32]
plt.xlabel("Days")
plt.ylabel("Temp")
plt.title("Day and Temp")
plt.plot(day,temp,color='red',linestyle='dotted')
plt.show()
```

```python
# Multiple graphs
day = [1,2,3,4,5,6,7]
temp = [30,33,34,32,28,27,32]
wind = [10,20,25,25,26,27,28]

plt.xlabel("Days")
plt.ylabel("Temp and wind")
plt.title("Day and Temp/wind")

plt.plot(day,temp,color='red',linestyle='dotted',label='Temp')  # plot as many as graph as you required with plt.plot for different x and y
plt.plot(day,wind,color='red',linestyle='-', label = 'Wind')

plt.legend(loc = "lower right")   # location for legend, default is 'best', else you can try 'upper right'
plt.grid()
plt.show()
```

### Bar Chart

```python
company = ['FB','Google','Nvidia','Amzn']
revenue = [100,120,95,118]

ypos = np.arange(len(company))  # converts 'FB','Google','Nvidia','Amzn' to 0,1,2,3

plt.xticks(ypos,company) # maps numbers to  company list

plt.xlabel("Company")
plt.ylabel("Revenue")
plt.title("Company Vs Revenue")

plt.bar(ypos, revenue)
plt.show()
```

```python
# Plotting two bar graphs

company = ['FB','Google','Nvidia','Amzn']
revenue = [100,120,95,118]
profit = [20,24,19,23]

ypos = np.arange(len(company))  # converts 'FB','Google','Nvidia','Amzn' to 0,1,2,3

plt.xticks(ypos,company) # maps numbers to  company list

plt.xlabel("Company")
plt.ylabel("Revenue")
plt.title("Company Vs Revenue-Profit")

plt.bar(ypos, revenue,label='Revenue')
plt.bar(ypos, profit,label='Profit')

plt.legend()
plt.show()
```

```python
# Plotting two bar graphs side by side

company = ['FB','Google','Nvidia','Amzn']
revenue = [100,120,95,118]
profit = [20,24,19,23]

xpos = np.arange(len(company))  # converts 'FB','Google','Nvidia','Amzn' to 0,1,2,3

plt.xticks(xpos,company) # maps numbers to  company list

plt.xlabel("Company")
plt.ylabel("Revenue")
plt.title("Company Vs Revenue-Profit")

plt.bar(xpos-0.2, revenue,width=0.4,label='Revenue')
plt.bar(xpos+0.2,profit,width=0.4,label='Profit')

plt.legend()
plt.show()
```

### Horizontal Bar Chart: Barh

```python
company = ['FB','Google','Nvidia','Amzn']
revenue = [100,120,95,118]

ypos = np.arange(len(company))  # converts 'FB','Google','Nvidia','Amzn' to 0,1,2,3

plt.yticks(ypos,company) # maps numbers to  company list

plt.xlabel("Revenue")
plt.ylabel("Company")
plt.title("Company Vs Revenue")

plt.barh(ypos, revenue)
plt.show()
```

```python
# Histogram for two sample

# Custom bins
blood_sugar_men = [85,90,100,110,110,112,115,178,155,165,135,145,120,125,127,130,135,138]
blood_sugar_women = [80,95,105,100,120,102,105,148,135,128,135,145,120,125,127,130,135,138]

plt.xlabel("Sugar range")
plt.ylabel("frequency")
plt.title("Men-Women Blood Sugar")

plt.hist([blood_sugar_men,blood_sugar_women],bins=[80,100,125,150], rwidth=0.95, label=['men','women'],color=['r','g'],cumulative=True)
plt.legend()
plt.show()
```

```python
# Histogram orientation horizontal

# Custom bins
blood_sugar_men = [85,90,100,110,110,112,115,178,155,165,135,145,120,125,127,130,135,138]
blood_sugar_women = [80,95,105,100,120,102,105,148,135,128,135,145,120,125,127,130,135,138]

plt.xlabel("Sugar range")
plt.ylabel("frequency")
plt.title("Men-Women Blood Sugar")

plt.hist([blood_sugar_men,blood_sugar_women],bins=[80,100,125,150], rwidth=0.95, label=['men','women'],color=['r','g'],cumulative=True, orientation='horizontal')
plt.legend()
plt.show()
```

### Pie Chart

```python
exp = [6500,2500,2000,8000]
exp_label = ['rent','groce','medi','fees']

plt.axis("equal")
plt.pie(exp,labels= exp_label, autopct='%0.2f%%')
plt.show()
```

```python
%matplotlib inline
exp = [6500,2500,2000,8000]
exp_label = ['rent','groce','medi','fees']

plt.axis("equal")
plt.pie(exp,labels= exp_label, autopct='%0.2f%%', explode=[0,0.3,0,0]) # explode takes out piece
plt.show()
```

### Save Chart to Image

```python
%matplotlib inline
exp = [6500,2500,2000,8000]
exp_label = ['rent','groce','medi','fees']

plt.axis("equal")
plt.pie(exp,labels= exp_label, autopct='%0.2f%%', explode=[0,0.3,0,0]) # explode takes out piece
#plt.show()

plt.savefig('piechart.jpg',bbox_inches='tight',pad_inches=0.5) # pad_inches: extra space all around image
```
